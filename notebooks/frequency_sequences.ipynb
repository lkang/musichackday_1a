Utilities to generate impulse sequences with specific frequencies


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gen_freq_impulse_sequence(freq, length, sample_rate=44100):
    #
    sequence = np.zeros(length)
    # (samples / cycle) = (samples / sec) / (cycles / sec)
    x = sample_rate / freq
    for s in range(length):
        if s % x == 0:
            sequence[s] = 100.0
    return sequence

def gen_freq_sin_sequence(freq, length, sample_rate=44100):
    radians = np.pi * 2 * freq
    numcycles = length/sample_rate
    total_radians = numcycles * radians
    x = np.linspace(0,total_radians,length)
    sequence = np.sin(x)
    return sequence
```

```python
testseq_impulse = gen_freq_impulse_sequence(2, 90000)

print(testseq_impulse)
```

```python
plt.plot(testseq_impulse)
plt.show()
```

```python
fft_result_impulse = np.fft.fft(testseq_impulse)

print(fft_result_impulse)

plt.plot(np.absolute(fft_result_impulse[:200]))
plt.show()
```

```python
testseq = gen_freq_sin_sequence(2, 90000)

print(testseq)
```

```python
plt.plot(testseq[:])
plt.show()
```
